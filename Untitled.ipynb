### Keep only the following features: precipitation, mean_temp, min_temp, max_temp, snow_depth

## 4 Impute any missing values.
Think about each feature and what imputation method makes sense. You may not be able to use the same method for every feature!

```python
df.isna().sum()
```

Interpolation is a technique in Python with which you can estimate unknown data points between two known data points. It is commonly used to fill missing values in a table or a dataset using the already known values.

```python
## interpolating temperature measures
df['mean_temp'] = df['mean_temp'].interpolate()
df['max_temp'] = df['max_temp'].interpolate()
df['min_temp'] = df['min_temp'].interpolate()
```

```python
df['mean_temp']
```

```python
df.isna().sum()
```

```python
# filling preciptation/snow with 0
df['precipitation'] = df['precipitation'].fillna(0)
df['snow_depth'] = df['snow_depth'].fillna(0)
```

```python
df.isna().sum()
```

## 5. Answer the following questions using .idxmax() and .idxmin()

Print statements, and the .strftime() method. Use f-strings to print the following statements (with the blanks filled in):

a) "The lowest temperature of {max temp} degrees occurred on {date as "MonthName 2-digit-day, 4-digit year. } which was a {day of the week}."
​

```python
lowest_temperature = df['min_temp'].min()
lowest_temperature
```

```python
lowest_temperature_date = df['min_temp'].idxmin()
lowest_temperature_date 
```

```python
lowest_temperature_date.strftime('%B %d, %Y')
```

```python
print(f"The lowest temperature of {lowest_temperature} degrees occurred on {lowest_temperature_date.strftime('%B %d, %Y')} which was a {lowest_temperature_date.day_name()}.")
```
